```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
House_dataset = pd.read_csv(r"F:\project.py\data set predection\House price predection\Housing.csv")
House_dataset
```

```python
x=House_dataset[["area","bedrooms",	"bathrooms","stories","mainroad","guestroom","basement","hotwaterheating","airconditioning","parking","prefarea","furnishingstatus"]]
x
```

```python
y=House_dataset["price"]
y
```

```python
x["mainroad"] = x["mainroad"].map({"no": 0, "yes": 1})
x["guestroom"] = x["guestroom"].map({"no": 0, "yes": 1})
x["basement"] = x["basement"].map({"no": 0, "yes": 1})
x["hotwaterheating"] = x["hotwaterheating"].map({"no": 0, "yes": 1})  # Fixed mapping
x["airconditioning"] = x["airconditioning"].map({"no": 0, "yes": 1})
x["prefarea"] = x["prefarea"].map({"no": 0, "yes": 1})
x["furnishingstatus"] = x["furnishingstatus"].map({"furnished": 0, "semi-furnished": 1, "unfurnished": 2})


x = x.fillna(x.select_dtypes(include=['number']).mean())
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
area=int(input("Enter the area(1000/10000):"))
bedrooms=int(input("Enter the bedrooms(1/4):"))
bathrooms=int(input("Enter the bathrooms(1/4):"))
stories	=int(input("Enter the stories(1/4):"))
mainroad=input("Enter the mainroad(yes/no)")
guestroom=input("Enter the guestroom(yes/no):")	
basement=input("Enter the basement(no/yes):")
hotwaterheating=input("Enter the hotwaterheating(no):")
airconditioning=input("Enter the airconditioning(no/yes):")
parking=int(input("Enter the parking(0/3):"))
prefarea=input("Enter the prefarea(no/yes):")
furnishingstatus=input("Enter the furnishingstatus(furnished/unfurnished/semi-furnished) ")

mainroad = 1 if mainroad.lower() == "yes" else 0
guestroom = 1 if guestroom.lower() == "yes" else 0
basement = 1 if basement.lower() == "yes" else 0
hotwaterheating = 1 if hotwaterheating.lower() == "yes" else 0
airconditioning = 1 if airconditioning.lower() == "yes" else 0
prefarea = 1 if prefarea.lower() == "yes" else 0
furnishingstatus = {"furnished": 0, "semi-furnished": 1, "unfurnished": 2}.get(furnishingstatus.lower(), -1)

# Prepare input data for prediction
user_data = [[area, bedrooms, bathrooms, stories, mainroad, guestroom, basement, 
              hotwaterheating, airconditioning, parking, prefarea, furnishingstatus]]

# Predict house price
predicted_price = model.predict(user_data)
print("Predicted House Price:", predicted_price[0])

```
